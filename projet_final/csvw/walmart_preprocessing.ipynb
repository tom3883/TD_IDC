The goal of this notebook is to process the csv walmart dataset to reduce its size (rows and columns).

Original dataset is not included to its large size but can be found here : https://www.kaggle.com/datasets/thedevastator/product-prices-and-sizes-from-walmart-grocery?resource=download

```python
import pandas as pd
```

```python
df = pd.read_csv("./WMT_Grocery_202209.csv")

df.head()
```

```python
# Removing unwanted columns

colums_to_remove = ["SHIPPING_LOCATION", "DEPARTMENT", "PRODUCT_SIZE", "PRICE_RETAIL", "SUBCATEGORY", "BREADCRUMBS", "SKU", "PRODUCT_URL", "PROMOTION", "RunDate", "tid"]

df = df.drop(columns=colums_to_remove)
```

```python
df
```

```python
# remove brand name from the product name
def remove_brand_name(product_name, brand_name):
    if pd.notna(brand_name) and brand_name in product_name:
        product_name = product_name.replace(brand_name, '').strip()
        
    # remove everything after the comma
    comma = product_name.find(',')
    if comma != -1:
        product_name = product_name[:comma].strip()
    
    return product_name

df['PRODUCT_NAME'] = df.apply(lambda row: remove_brand_name(row['PRODUCT_NAME'], row['BRAND']), axis=1)

df
```

```python
# changing column order
df = df[['index', 'PRODUCT_NAME', 'BRAND', 'CATEGORY', 'PRICE_CURRENT']]
cols = list(df.columns.values)
print(cols)
```

```python
#skip every 2nd line
df = df.iloc[::17]

df
```

```python
# Download the filtered dataset
df.to_csv('./walmart_products_reduced.csv', index=False)
```
